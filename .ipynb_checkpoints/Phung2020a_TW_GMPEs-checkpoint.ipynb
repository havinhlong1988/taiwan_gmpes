```python
'''
Function of GPMEs evaluated by Phung et al 2020a for subduction zone
Function form: lnPSA = a1 + Fmag(M,theta) + Fdist(M,R,theta) + a7Feve + Fdep(Ztor,theta)
+ Fsite(Vs30,theta) + Fbasin(Z1.0,theta) + eta + epsilon
'''
def Fmag(M,theta):
    import pandas as pd
    '''
    Functional form of the magnitude term
    '''
    d = pd.read_csv('Phung2020a.csv',header=0,delim_whitespace=True,
                    names=['T','a1','a4','a5','a13','Mref'])
    Fmag = [];
    for i,T in enumerate(d['T']):
        if (M <= d['Mref'][i]):
            Fma = d['a4'][i]*(M-d['Mref'][i]) + d['a13'][i]*(10 - M)**2
        else:
            Fma = d['a5'][i]*(M-d['Mref'][i]) + d['a13'][i]*(10 - M)**2
        Fmag.append(Fma)
    Fmag = pd.DataFrame(Fmag); # Convert to data frame
    return Fmag
```
